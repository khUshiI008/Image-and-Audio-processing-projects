```python
# Folder containing your audio files
audio_folder = r"c:\Users\k2664"

```

```python
# Get a list of all files in the folder
all_files = os.listdir(audio_folder)

```

```python
# Filter files to only include those with a .wav extension
audio_files = [file for file in all_files if file.endswith(".wav")]

```

```python
# Output folder for processed files
output_folder = "processed_audio_folder"
os.makedirs(output_folder, exist_ok=True)

```

```python
# Parameters for processing
sampling_rate_fixing = 44100
duration_fixing = 12.0
noise_red_fixing = 0.6

```

```python
# Process each audio file
for audio_file in audio_files:
    # Construct full path to input file
    input_file_path = os.path.join(audio_folder, audio_file)

    # Check if the file exists
    if not os.path.exists(input_file_path):
        print(f"File not found: {input_file_path}")
        continue

    # Load the audio file using pydub
    audio = AudioSegment.from_file(input_file_path)

    # Fix the audio properties
    audio = audio.set_frame_rate(sampling_rate_fixing)
    audio = audio.set_channels(1)  # Set to mono
    audio = audio[:int(duration_fixing * 1000)]  # Trim to desired duration

    # Apply noise reduction (if needed)
    if noise_red_fixing > 0.0:
        audio = audio - noise_red_fixing

    # Construct full path to output file
    output_file_name = f"processed_{os.path.basename(audio_file)}"
    output_file_path = os.path.join(output_folder, output_file_name)

    # Export the processed audio
    audio.export(output_file_path, format="wav")

    print(f"Processed {audio_file} and saved as {output_file_path}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import read
audio_files = ["Day-1 (Morning).wav", "Day-1 (Afternoon).wav", "Day-1(Evening).wav", "Day-1 (Night).wav"]
audio_mads = []
for audio_file in audio_files:
    fs, audio_data = read(audio_file)
    audio_data = np.array(audio_data)
    audio_mean = np.mean(audio_data)
    audio_diff = np.abs(np.subtract(audio_data, audio_mean))
    audio_mad = np.mean(audio_diff)
    audio_mads.append(audio_mad)
plt.bar(audio_files, audio_mads)
plt.xlabel('Audio file')
plt.ylabel('Mean average difference')
plt.show()
```

```python
# Define the day1 audio files
morning_audio_files = [
    '/content/processed_audio_folder/processed_Day-1 (Morning).wav',
    '/content/processed_audio_folder/processed_Day-1 (Afternoon).wav',
    '/content/processed_audio_folder/processed_Day-1(Evening).wav',
    '/content/processed_audio_folder/processed_Day-1 (Night).wav',
]

# Define the Mean_of_audio function
def Mean_of_audio(audio_file_path):
    audio_data, sr = librosa.load(audio_file_path, sr=None)
    return np.mean(audio_data)

# Calculate the means of all morning audio files
audio_means = [Mean_of_audio(audio_file) for audio_file in morning_audio_files]

# Calculate the differences between means
mean_differences = [audio_means[m] - audio_means[m - 1] for m in range(1, len(audio_means))]

# Plotting chart
plt.figure(figsize=(10, 5))
plt.plot(range(1, len(audio_means)), mean_differences, marker='o', linestyle='-')
plt.title('Average Mean Differences Across Morning Audios')
plt.xlabel('Audio File Index')
plt.ylabel('Mean Difference')
plt.grid(True)
plt.show()
```
